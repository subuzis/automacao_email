# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: lobat.dev@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
#Importar as bibliotecas (previamente instaladas)

import pyautogui
import pyperclip
import time

pyautogui.PAUSE = 1

#Passo 1 - entrar no sistema (link: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing)
#com o navegador google chrome aberto!!ele vai abrir uma nova aba
pyautogui.hotkey('ctrl', 't')

#pyautogui.press('win') --abrir navegador com o Python

link = 'https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing'
pyperclip.copy(link)  ### tem que usar para escrever textos com caracteres especiais que não são tratados com o pyautogui
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')

#Passo 2 - Navegar no sistema - abrir a pasta onde está o arquivo...se for um link direto não precisa
time.sleep(5) ##espera 5 segundos
pyautogui.click(x=408, y=299, clicks=2)  ##duplo click


#Passo 3 - Baixar o arquivo desejado (nesse caso o arquivo de vendas)
time.sleep(3)##espera 3 segundos
pyautogui.click(x=391, y=419) ###clicar no arquivo
time.sleep(3)
pyautogui.click(x=1165, y=195) ###clicar nos 3 pontinhos
time.sleep(3)
pyautogui.click(x=922, y=586) ###clicar no download
time.sleep(7)##esperar fazer o download
pyautogui.press('enter')
time.sleep(7)






```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
#Passo 4 - Calcular faturamento e quantidade de produtos vendidos
import pandas as pd

tabela = pd.read_excel(r"C:\....\Vendas - Dez.xlsx") #caminho do arquivo
faturamento = tabela["Valor Final"].sum()
quantidade = tabela["Quantidade"].sum()
display(tabela)
```

### Vamos agora enviar um e-mail pelo gmail

```python
#Passo 5 - Enviar um e-mail com as informações coletadas
pyautogui.hotkey('ctrl', 't')
link = 'https://mail.google.com'
pyperclip.copy(link)  ### tem que usar para escrever textos com caracteres especiais que não são tratados com o pyautogui
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')
time.sleep(7)

##Escrever e-mail

#clicar no botão escrever
pyautogui.click(x=72, y=204)
time.sleep(3)
#Digitar o e-mail do destinatátio
pyautogui.write('emaildestinatario')##coloque o e-mail do destinatário
time.sleep(2)
pyautogui.press('tab') ##escolhe e-mail
time.sleep(3)
pyautogui.press('tab') #passa para campo assunto

#Digitar assunto
assunto = ('Relatório de Vendas')
pyperclip.copy(assunto)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('tab') #passa para o corpo do e-mail

#Escrever no corpo do e-mail
texto = f"""
Prezados, bom dia,

O faturamento de ontem foi de R${faturamento:,.2f}

A quantidade de produtos vendidos ontem foi de {quantidade:,}
"""
pyautogui.write(texto)
time.sleep(2)
#Enviar e-mail
pyautogui.hotkey('ctrl', 'enter') ###atalho do gmail
```
